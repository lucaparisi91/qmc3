```python
def readPimc(folder):
    particleFile=os.path.join(folder,"particles.dat")
    settingsFile=os.path.join(folder,"description.dat")
    conf=pd.read_csv(particleFile,sep=" ",index_col=False)
    return conf
conf=readPimc(folder)
```

```python
def plotConfigurationParticleView(data):
        data=data.sort_values(by="time")
        data["particle"]=data["particle"].astype(str)
        graphs=[]
        for particle,particleData in data.groupby("particle"):
            graphs.append(go.Scatter3d( x=particleData['x'], y=particleData['y'], z=particleData['z'] ,name=str(particle)))
        fig=go.Figure(data=graphs)
        return fig
    

def plotConfigurationTimeView(data):
        graphs=[]
        data=data.sort_values(by="time")
        graphs.append(go.Scatter3d( x=data['x'], y=data['y'], z=data['z'] 
                                       , mode='markers',
                                       marker=dict(
                                            size=12,
                                            color=data["time"],
                                            colorscale='Viridis',
                                            opacity=0.8,
                                           colorbar=dict(
            title="time"
        )
    )
                                      ))
        fig=go.Figure(data=graphs)
        return fig

    
    
def plotConfiguration(data,view="particle"):
    if view=="particle":
        return plotConfigurationParticleView(data)
    else:
        if view == "time":
            return plotConfigurationTimeView(data)
    
plotConfiguration(conf,view="particle")
```
